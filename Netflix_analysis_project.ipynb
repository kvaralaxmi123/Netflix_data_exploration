#**Bivarient Plot**

##3.Analysis of actors/directors of different types of shows/movies.

```python
df.groupby('type')[['director','cast']].value_counts()
```

```python
df = df.dropna(subset=['cast'])
```

```python
df['cast'] = df['cast'].str.split(', ')
df_exploded = df.explode('cast')
```

```python
actor_counts = df_exploded.groupby(['type', 'cast']).size().reset_index(name='count')
top_actors_movies = actor_counts[actor_counts['type'] == 'Movie'].nlargest(10, 'count')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_actors_movies, x='count', y='cast', palette='viridis')
plt.title("Top 10 Actors in Netflix Movies")
plt.xlabel("Number of Appearances")
plt.ylabel("Actor")
plt.tight_layout()
plt.show()
```

```python
df['director'] = df['director'].str.split(', ')
df_exploded = df.explode('director')
```

```python
director_counts = df_exploded.groupby(['type', 'director']).size().reset_index(name='count')

top_10_movie_directors = director_counts[director_counts['type'] == 'Movie']\
 .nlargest(10, 'count')
top_10_show_directors = director_counts[director_counts['type'] == 'TV Show']\
    .nlargest(10, 'count')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

sns.barplot(data=top_10_movie_directors, y='director', x='count', ax=axes[0], palette="Blues_d")
axes[0].set_title("Top 10 Directors - Movies")
axes[0].set_xlabel("Number of Movies")
axes[0].set_ylabel("Director")

sns.barplot(data=top_10_show_directors, y='director', x='count', ax=axes[1], palette="Greens_d")
axes[1].set_title("Top 10 Directors - TV Shows")
axes[1].set_xlabel("Number of TV Shows")
axes[1].set_ylabel("Director")

plt.tight_layout()
plt.show()
```

##5.Understanding what content is available in different countries

```python
df['country'].value_counts().reset_index(name='count')
```

```python
grouped_df = df.groupby(['country', 'type']).size().reset_index(name='count')
```

```python
top_countries = (grouped_df.groupby('country')['count'].sum().sort_values(ascending=False).head(10).index)
grouped_df = grouped_df[grouped_df['country'].isin(top_countries)]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=grouped_df, x='country', y='count', hue='type')
plt.title('Number of Movies and TV Shows by Country(Top-10 countries)', color = 'purple')
plt.ylabel('Count')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#6. best time to launch a Tv show

```python
df.head()
```

```python
tv_shows = df[df['type'] == 'TV Show'].copy()
```

```python
tv_shows
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['month'] = df['date_added'].dt.month
df['weekday'] = df['date_added'].dt.day_name()
```

```python
best_month = df['month'].value_counts().idxmax()
best_weekday = df['weekday'].value_counts().idxmax()

print(f"Best month to launch a TV show: {best_month}")
print(f"Best day to launch a TV show: {best_weekday}")
```

```python
heatmap_data = df.groupby(['weekday', 'month']).size().unstack(fill_value=0)

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title("TV Shows Added by Weekday and Month")
plt.xlabel("Month")
plt.ylabel("Weekday")
plt.show()
```
